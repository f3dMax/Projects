```python
# Create a variable called response that stores the response from the api call using the requests library. Remember .get is a method that takes a url as an argument and returns the response from the api call.
response = requests.get(url)
# Create a variable called data that stores the json data from the response.
data = response.json()

# Print the data variable.
print (data)

```

```python
# Time to extract the data we want from the JSON data. We will extract the temperature, the humidity, and the Weather description. We will store each of these in a variable. Using the Jupyter Tab in the terminal below you can see the structure of the JSON data. 
temp = data["main"]["temp"]
humidity = data["main"]["humidity"]
description = data["weather"][0]["description"]


# For example the data["main"] stores the main data. The data["main"]["temp"] stores the temperature. 
print(data["main"])

# Create a variable called temp that stores the temperature from the data.
temp_f = round((temp - 273.15) * 9/5 + 32, 2)

# Create a variable called humidity that stores the humidity from the data.


# The data["weather"] stores the weather data.
print(data["weather"])

# Create a variable called description that stores the description from the data. Notice the data["weather"] is a list. We can access the first element of the list by using data["weather"][0]. Then access the description using another set of brackets.

# Print the temp, humidity, and description variables.
# The temp variable is in Kelvin. We need to convert it to Fahrenheit. The formula to convert Kelvin to Fahrenheit is (K - 273.15) * 9/5 + 32. Create a variable called temp_f that stores the temperature in Fahrenheit.
# Note: replace K with the temp variable.
# Print the temp_f variable. Round the number to 2 decimal places.

print(f"Temperature (in Fahrenheit): {temp_f}F")


print(f"Temperature (in Kelvin): {temp}")
print(f"Humidity: {humidity}%")
print(f"Weather Description: {description}")
```
